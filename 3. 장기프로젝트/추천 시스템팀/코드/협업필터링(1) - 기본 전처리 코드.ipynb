```python
from collections import Counter

def show_most_frequent(column):
    counter = Counter(column)
    most_common = counter.most_common()

    for value, count in most_common:
        print(f'{value}: {count}')

show_most_frequent(df['state'])

# califonia, texas, ontario, missouri, florida
```

```python
def extract_top_attributes(column):
    counter = Counter(column)
    total_count = sum(counter.values())
    sorted_counts = sorted(counter.items(), key=lambda x: x[1], reverse=True)

    cumulative_percentage = 0
    extracted_attributes = []

    for value, count in sorted_counts:
        cumulative_percentage += count / total_count
        extracted_attributes.append(value)

        if cumulative_percentage >= 0.6:

            break

    return extracted_attributes
```

```python
top_attributes = extract_top_attributes(df['state'])
print(top_attributes)
# 60% 기준으로 잘라서 범주화하고 나머지는 'others'로 분류
```

```python
test=[]
for j in range(len(top_attributes)):
  if len(top_attributes[j]) >= 4:
    test.append(top_attributes[j])
```

```python
for i in range(len(df)):
  if df['state'][i] not in test:
    df['state'][i] = 'others'
```

# 출판연도 처리

```python
df['Year-Of-Publication'].describe()
```

```python
Counter(df['Year-Of-Publication'])
```

```python
df=df[df['Year-Of-Publication'] != -1.0] # 7913개의 Outlier 제거
```

```python
column_name = 'Year-Of-Publication'
has_null = df[column_name].isnull().any()

if has_null:
    # 결측치가 있는 행 제거
    df = df.dropna(subset=[column_name])
```

```python
df = df.reset_index(drop=True)
```

```python
for i in range(len(df)):
  if df['Year-Of-Publication'][i] < 1980.0:
    df['Year-Of-Publication'][i] = '-1980'
  elif 1980.0 <= df['Year-Of-Publication'][i] < 1985.0:
    df['Year-Of-Publication'][i] = '1980-1984'
  elif 1985.0 <= df['Year-Of-Publication'][i] < 1990.0:
    df['Year-Of-Publication'][i] = '1985-1989'
  elif 1990.0 <= df['Year-Of-Publication'][i] < 1995.0:
    df['Year-Of-Publication'][i] = '1990-1994'
  elif 1995.0 <= df['Year-Of-Publication'][i] < 2000.0:
    df['Year-Of-Publication'][i] = '1995-1999'
  else:
    df['Year-Of-Publication'][i] = '2000+'
```

```python
df['Year-Of-Publication'] = df['Year-Of-Publication'].astype('category')
```
