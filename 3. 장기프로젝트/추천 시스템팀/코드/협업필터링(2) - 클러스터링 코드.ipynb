# 유저별로 한 개의 클러스터 지정(최빈값)

```python
from scipy.stats import mode
```

```python
def replace_with_mode(df):
    result = df.copy()
    result['clusters'] = result.groupby('User-ID')['clusters'].transform(lambda x: mode(x)[0][0])
    return result
```

```python
df_1 = replace_with_mode(df)
```

```python
df_1['clusters'].value_counts()
```

```python
cluster0=df_1[df_1['clusters']==0]
cluster1=df_1[df_1['clusters']==1]
cluster2=df_1[df_1['clusters']==2]
cluster3=df_1[df_1['clusters']==3]
cluster4=df_1[df_1['clusters']==4]
cluster5=df_1[df_1['clusters']==5]
cluster6=df_1[df_1['clusters']==6]
cluster7=df_1[df_1['clusters']==7]
```

```python
df_1.to_csv("df_cluster.csv")
cluster0.to_csv('cluster0.csv')
cluster1.to_csv('cluster1.csv')
cluster2.to_csv('cluster2.csv')
cluster3.to_csv('cluster3.csv')
cluster4.to_csv('cluster4.csv')
cluster5.to_csv('cluster5.csv')
cluster6.to_csv('cluster6.csv')
cluster7.to_csv('cluster7.csv')
```
