```python
sample = pd.DataFrame(cons1.loc['row']['GeoInfoParkAndPrivateLandWGS'])
sample.head()
```

```python
trees_data = sample.copy()
num = 1

#API KEY
my_key = '45657051756b6d61313038514d71504a'

while True:
    num = num + 1000
  # 샘플 url 변형해서 사용
    endpoint=f'http://openAPI.seoul.go.kr:8088/{my_key}/json/GeoInfoParkAndPrivateLandWGS/{num}/{num+999}'

  # api 호출
    resp=requests.get(endpoint)
    print(resp.status_code)
    print(resp.text[0:200])
    try:
    # json 파일로 가져왔기때문에, json() 함수로 data에 넣어주기
        data=resp.json()
        temp_ = pd.DataFrame(data)
        temp_trees = pd.DataFrame(temp_.loc['row']['GeoInfoParkAndPrivateLandWGS'])
        print(num, len(temp_trees))
        trees_data = pd.concat([trees_data, temp_trees])
    except:
        break

trees_data.info()
```
