문자형 컬럼을 수치형으로 변환하는 중에, '정량한계미만'이라는 측정값 때문에 오류가 발생함.

```python
data['클로로필-a(Chlorophyll-a)'] = pd.to_numeric(data['클로로필-a(Chlorophyll-a)'])
```

```python
interested = ['클로로필-a(Chlorophyll-a)', '용존총질소(DTN)',
       '용존총인(DTP)', '분원성대장균군', '질산성질소(NO3-N)', '부유물질(SS)', '총대장균군',
       '총유기탄소(TOC)', '총인(T-P)']
for col in interested:
    data.loc[data[col] == '정량한계미만', col] = '0'
```

### 3) 요인분석

```python
!pip install factor_analyzer
```

```python
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity, calculate_kmo
```

```python
df = data.drop([' 수계 명', ' 중권역 명', ' 측정소 명'], axis=1)

chi_square_value, p_value = calculate_bartlett_sphericity(df)
print(chi_square_value, p_value)
```

바틀렛의 구면성 검정 결과, 변수간 상관성이 없다는 귀무가설을 기각하여,  
요인분석이 의미를 가짐을 알 수 있다.

```python
kmo_all, kmo_model = calculate_kmo(df)
kmo_model
```

KMO(Kaiser-Meyer-Olkin) 검정 결과, KMO 값이 0.6과 1 사이이므로,  
요인 분석이 적합함을 알 수 있다.

```python
# 요인 수 선택
fa = FactorAnalyzer(n_factors=16, rotation=None)
fa.fit(df)

ev, v = fa.get_eigenvalues()
ev
```

```python
plt.scatter(range(1, df.shape[1]+1), ev)
plt.plot(range(1, df.shape[1]+1), ev)
plt.grid()
plt.show()
```

eigenvalue 기준으로는 5개, 그래프 기울기 기준으로는 7개의 요인으로 나누어주는 것이 최선이다.

```python
# 요인 7개
fa = FactorAnalyzer(n_factors=7, rotation='varimax')
fa.fit(df)
efa_result = pd.DataFrame(fa.loadings_, index=df.columns)
efa_result
```

```python
plt.figure(figsize=(6,8))
sns.heatmap(efa_result, cmap='Blues', annot=True, fmt='.2f')
```

7개의 요인 중 요인 0은 TOC에 대한 높은 요인 적재량을 보임.

결과적으로 같은 요인에 있는 BOD, Chl-a, COD, SS, (T-P)가 TOC에 많은 영향을 주는 것을 알 수 있다.

### 4) 회귀분석

```python
y = df['총유기탄소(TOC)']
X = df[df.drop('총유기탄소(TOC)', axis=1).columns]
```

```python
from sklearn.preprocessing import RobustScaler

robust_scaler = RobustScaler()

# X의 모수 분포 저장
robust_scaler.fit(X)

# X 스케일링
X_scaled = robust_scaler.transform(X)
```

```python
X_vif = pd.DataFrame(X_scaled)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
vif['features'] = X_vif.columns
vif = vif.sort_values('VIF Factor').reset_index(drop=True)
vif
```

스케일링만으로는 다중공선성의 문제가 해결되지 않았다.

```python
from statsmodels.formula.api import ols

model = ols("y ~ X_scaled", data = df).fit()
model.summary()
```

기본적인 회귀분석 결과,

BOD, Chlorophyll-a, COD, 용존총질소(DTN), 용존총인(DTP), 전기전도도(EC),  
질산성질소(NO3-N), SS, 수온, T-N, T-P 컬럼의 p값이 유의함을 알 수 있다.

또한 상대적으로 영향을 많이 미치는 변수는  
COD(+), 용존총질소(DTN)(+), 수온(+), T-N(-) 이다.

단, 다중공선성이 존재하는 것으로 보인다.
