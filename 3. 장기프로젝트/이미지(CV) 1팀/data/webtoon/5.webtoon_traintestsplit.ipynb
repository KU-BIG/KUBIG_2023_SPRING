```python
import os
import random
import shutil

source_folder = '/content/drive/MyDrive/23-1 CV프로젝트 1팀/datasets_final/trainA/' # 학습 이미지가 모두 저장되어있는 폴더
validation_folder = '/content/drive/MyDrive/23-1 CV프로젝트 1팀/datasets_final/testA/' # val(또는 test) 폴더

validation_percentage = 0.2  # 8:2

# Get a list of all the image files in the source folder
image_files = [file for file in os.listdir(source_folder) if file.endswith('.jpg')]

# Calculate the number of images to move to the validation folder
num_images = len(image_files)
num_validation = int(num_images * validation_percentage)

# Randomly select the images to move
validation_images = random.sample(image_files, num_validation)

# Move the selected images to the validation folder
for image in validation_images:
    source_path = os.path.join(source_folder, image)
    destination_path = os.path.join(validation_folder, image)
    shutil.move(source_path, destination_path)

print(f"{num_validation} images moved to the validation folder.")
```
