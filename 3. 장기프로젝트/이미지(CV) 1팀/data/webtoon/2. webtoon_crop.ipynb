```python
save_path = '/content/drive/MyDrive/23-1 CV프로젝트 1팀/add_webtoon/crop/' # 저장 폴더
image_path = '/content/drive/MyDrive/23-1 CV프로젝트 1팀/add_webtoon/webtoon/' # 이미지 들어있는 폴더
```

```python
# 드라이브에 웹툰 제목으로 crop 이미지 폴더 생성(한번만 돌리기!)

for webtoon in webtoon_title:
  os.mkdir(os.path.join(save_path,  webtoon))
```

```python
for webtoon in webtoon_title:
  first_path = image_path + webtoon + '/'

  cnt = 0
  print(webtoon + " 크롭 시작!")

  for i in range(1, 21): # 1화부터 20화까지
    print(str(i) + "화 crop 시작!")
    path = first_path + str(i)
    file_list = os.listdir(path)
    file_list_py = [file for file in file_list if file.endswith('.jpg')]

    for name in file_list_py:
      image = cv2.imread(path + '/' + name)
      img = Image.fromarray(image)  # RGB image
      faces = animeface.detect(img)

      for face in faces:
        x, y, width, height = face.face.pos.x, face.face.pos.y, face.face.pos.width, face.face.pos.height
        cropped_image = image[y:y+height, x:x+width]
        cv2.imwrite(save_path + webtoon + '/' + "cropped_image" + str(cnt) + ".jpg", cropped_image)
        cnt = cnt + 1

    print(str(i) + "화 crop 완료!")
  print(webtoon + " 크롭 완료!")
```
